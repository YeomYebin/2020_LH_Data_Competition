## 해결과제 1 최종 결과 어린이 교통사고 위험지역 시각화

```python
import folium
import json
import pandas as pd
import numpy as np
import geopandas as gpd
import fiona
import matplotlib as plt
```

```python
#지도 만들기
center = [37.165397, 127.048396]
m = folium.Map(location=center, zoom_start=13, tiles='cartodb positron')
```

```python
# 결과 불러오기 
danger_score=gpd.read_file('result1_data.geojson')
final_result1=gpd.read_file('analysis1_result.geojson')
# 법정동 경계
data33 = gpd.read_file('33.오산시_법정경계(읍면동).geojson')
```

```python
# 최종 위험지역 결과
final_result1
```

```python
# 위험도 점수 격자
gridmap= folium.Map(location=center, zoom_start=13 ,tiles='cartodb positron')

folium.Choropleth(
    geo_data=danger_score,
    data=danger_score,
    columns=('gid', 'danger_score'),
    key_on='feature.properties.gid',
    fill_color='RdPu',
    legend_name='위험도 점수',
    line_opacity = .2
).add_to(gridmap)

# 맵 m을 출력합니다.
gridmap
gridmap.save('gridmap.html')
```

```python
style = {'fillOpacity':0.1, 'color': '#FFA7A7'}
folium.GeoJson(data=data33['geometry'], style_function=lambda x:style).add_to(m)
m
```

```python
for i in range(len(final_result1)):
    folium.Circle(
        location=[final_result1.iloc[i]['lat'], final_result1.iloc[i]['lon']],
        popup=final_result1.iloc[i]['시설명'] ,
        radius=20,
        color='crimson', fill=True, fill_color='crimson'
    ).add_to(m)
m
```
